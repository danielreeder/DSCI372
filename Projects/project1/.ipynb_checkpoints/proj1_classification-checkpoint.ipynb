```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import auc, roc_curve
```

```python
class LogisticRegression(object):
    def __init__(self, X_train, y_train, X_test, y_test, learn_rate=0.001, n_iters=500):
        if len(X_train.shape) == 1:
            self.X_train = np.hstack((np.array([1]), X_train))
        else:
            self.X_train = np.hstack((np.ones((X_train.shape[0], 1)), X_train))
        self.y_train = y_train

        if len(X_test.shape) == 1:
            self.X_test = np.hstack((np.array([1]), X_test))
        else:
            self.X_test = np.hstack((np.ones((X_test.shape[0], 1)), X_test))
        self.y_test = y_test

        self.learn_rate = learn_rate
        self.n_iters = n_iters 

        # Initialize model weights
        self.weight = np.zeros((self.X_train.shape[1]))

        self.accuracy_train = []  # To keep track of accuracy on the training data during gradient ascent process
        self.accuracy_test = []  # To keep track of accuracy on the test data during gradient ascent process

        self.log_likelihood_train = []  # To keep track of average log likelihood on the training data during gradient ascent process
        self.log_likelihood_test = []  # To keep track of average log likelihood on the test data during gradient ascent process

    def predict(self, samples, threshold=0.5):
        return self.predict_prob(samples) >= threshold

    def predict_prob(self, samples):
        # Todo: Compute and return the predicted probability of each sample having an income >= 50K
        # probabilities = np.zeros(samples.shape[0])
        probabilities = 1 / (1 + np.exp(-self.weight.dot(samples.T)))
        return probabilities
            
    def compute_gradient(self, ):
        # Todo: compute and return gradient of the average log likelihood on the training set
        i_range = np.arange(len(self.weight))
        j_range = np.arange(len(self.X_train))
        num_samples = len(self.X_train)
        grads = np.zeros(len(self.weight))
        
        y = learner.y_train
        y0 = 1 - y
        X = learner.X_train
        w = learner.weight
        z = w.dot(X.T)
        sig = 1 / (1 + np.exp(z))
        
        for i in i_range:  
            grads[i] = np.mean((y * (X[:, i] * sig)) + (y0 * -((X[:, i] * (np.exp(z))) * sig)))
        return grads
        
    def gradient_ascent(self, ):
        # Todo: Run gradient ascent to learn model weights
        # Todo: Store the average log likelihood and the prediction accuracy 
        #       on the training and test sets after every gradient ascent iteration
        for i in np.arange(self.n_iters):
            grads = self.compute_gradient()
            
            self.weight = self.weight + (self.learn_rate * grads)
            
            train_probs = self.predict_prob(self.X_train)
            test_probs = self.predict_prob(self.X_test)

            # getting predictions here instead of using self.predict() to optimize performance
            predictions_train = train_probs >= 0.5
            predictions_test = test_probs >= 0.5

            y_train = self.y_train
            y0_train = 1 - y_train

            y_test = self.y_test
            y0_test = 1 - y_test
            
            ave_logL_train = np.mean(np.log((y_train * train_probs) + (y0_train * (1 - train_probs))))
            ave_logL_test = np.mean(np.log((y_test * test_probs) + (y0_test * (1 - test_probs))))
            
            prediction_accuracy_train = np.mean(predictions_train == self.y_train)
            prediction_accuracy_test = np.mean(predictions_test == self.y_test)

            self.log_likelihood_train += [ave_logL_train]
            self.log_likelihood_test += [ave_logL_test]
    
            self.accuracy_train += [prediction_accuracy_train]
            self.accuracy_test += [prediction_accuracy_test]
        
```

```python
def preprocess_data(X_train, X_test):
    # encode categorical features using OneHotEncoder
    categorical_features = X_train.select_dtypes(include="object").columns
    t = [('encoder', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_features)]
    col_transform = ColumnTransformer(transformers=t)
    col_transform.set_output(transform="pandas")
    encoded_train = col_transform.fit_transform(X_train)
    encoded_test = col_transform.fit_transform(X_test)

    train = pd.concat([X_train, encoded_train], axis=1)
    train.drop(categorical_features, axis=1, inplace=True)
    test = pd.concat([X_test, encoded_test], axis=1)
    test.drop(categorical_features, axis=1, inplace=True)

    # standardize numerical features using StandardScaler()
    standard_scaler = StandardScaler()

    numeric_features = X_train.select_dtypes(include=["int64", "float64"]).columns
    
    standardized_train = pd.DataFrame(standard_scaler.fit_transform(train[numeric_features]), columns=numeric_features)
    standardized_test = pd.DataFrame(standard_scaler.fit_transform(test[numeric_features]), columns=numeric_features)

    train.drop(numeric_features, axis=1, inplace=True)
    train = pd.concat([train, standardized_train], axis=1)
    test.drop(numeric_features, axis=1, inplace=True)
    test = pd.concat([test, standardized_test], axis=1)

    # make sure train and test datasets have the same columns
    # any columns that are in train but not test are added to test as a column of zeros
    not_in_test = train.drop(columns=test.columns).columns
    test[not_in_test] = np.zeros([len(test), len(not_in_test)])

    test = test[train.columns]
                                 
    return train, test
```

```python
def plot_log_likelihood_performance(log_likelihood_train, log_likelihood_test):
    # Todo: plot the average log likelihood curves of the training and test sets
    #      during training process
    fig, ax1 = plt.subplots()

    ax1.plot(log_likelihood_train, c='r', label="logL train")
    ax1.plot(log_likelihood_test, c='b', label="logL test")
    ax1.set_title("Average Log Likelihood")
    ax1.set_xlabel("Iteration")
    ax1.set_ylabel("Average Log Likelihood")
    plt.legend()

    plt.savefig("plots/log_likelihood.jpg")
```

```python
def plot_accuracy_performance(accuracy_train, accuracy_test):
    # Todo: plot the prediction accuracy curves of the training and test sets
    #      during the training process
    fig, ax1 = plt.subplots()
    
    ax1.plot(accuracy_train, c='r', label="accuracy train")
    ax1.plot(accuracy_test, c='b', label="accuracy test")
    ax1.set_title("Accuracy")
    ax1.set_xlabel("Iteration")
    ax1.set_ylabel("Accuracy")
    plt.legend()

    plt.savefig("plots/accuracy.jpg")
    
```

```python
def plot_roc(X_train, y_train, X_test, y_test, learner):
    # Todo: plot the ROC curves 
    X_train = np.hstack((np.ones((X_train.shape[0], 1)), X_train))
    X_test = np.hstack((np.ones((X_test.shape[0], 1)), X_test))

    thresholds = np.arange(0, 1.05, 0.025)
    train_predictions = 1 * np.array([learner.predict(X_train, threshold=i) for i in thresholds])
    test_predictions = 1 * np.array([learner.predict(X_test, threshold=i) for i in thresholds])

    df_train = pd.DataFrame(train_predictions.T).apply(lambda x: 1 * (x == y_train), axis=0).T

    true_positives = df_train.apply(lambda x: sum(x * (y_train == 1)), axis=1)
    false_positives = df_train.apply(lambda x: sum((1 - x) * (y_train == 0)), axis=1)
    true_negatives = df_train.apply(lambda x: sum(x * (y_train == 0)), axis=1)
    false_negatives = df_train.apply(lambda x: sum((1 - x) * (y_train == 1)), axis=1)

    df_train["true_positive"] = true_positives / (true_positives + false_negatives)
    df_train["false_positive"] = false_positives / (false_positives + true_negatives)

    df_train = df_train[["true_positive", "false_positive"]]
    fig, ax = plt.subplots()

    ax.plot(df_train.false_positive, df_train.true_positive, c='b')
    ax.set_title("ROC Curve")
    ax.set_xlabel("False Positive Rate")
    ax.set_ylabel("True Positive Rate")

    plt.savefig("plots/roc.jpg")
    return
    
```

```python
# Load data
train_data = pd.read_csv('cleaned_train_data.csv')
test_data = pd.read_csv('cleaned_test_data.csv')
X_train = train_data.drop(['income'], axis=1, inplace=False)
y_train = train_data['income']
X_test = test_data.drop(['income'], axis=1, inplace=False)
y_test = test_data['income']
```

```python
# Preprocessing data
processed_X_train, processed_X_test = preprocess_data(X_train, X_test)
```

```python
# Run gradient ascent with Logistic Regression
learner = LogisticRegression(processed_X_train, y_train.values, processed_X_test, y_test.values,
                                 n_iters=250, learn_rate=0.75)

learner.gradient_ascent()
```

```python
# Plot prediction accuracy during training process
plot_accuracy_performance(learner.accuracy_train, learner.accuracy_test)
```

```python
# Plot Log Likelihood values during training process
plot_log_likelihood_performance(learner.log_likelihood_train, learner.log_likelihood_test)
```

```python
# Plot ROC curves
plot_roc(processed_X_train, y_train, processed_X_test, y_test, learner)
```
