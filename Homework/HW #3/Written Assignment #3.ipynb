```python
def get_counts(df, predictor, split):
    cols = np.array([predictor, split, "OK"])
    counts = df[cols].groupby(predictor).value_counts()
    
    return counts.reset_index()

def H(df, predictor):
    splits = np.delete(df.drop(columns="OK").columns.values, np.where(df.columns.values == predictor))
    # for split in splits:
    split = "Type"
    counts = get_counts(df, predictor, split).query("Price == '$' or Price == '$$'")
    vals = counts.value_counts("Price").index.values
    
    return counts
```

```python
one = H(orig_df, "Price").query("Price == '$'")
tot = one["count"].sum()

split = one.groupby("OK").sum()["count"]
rats = split / tot
-(sum(rats * np.log(rats)))
cas_split = one.query("Type == 'CAS'").groupby("OK").sum()["count"]
cas_split
fac = sum(cas_split) / tot
rats = cas_split / tot
-(sum(fac * (rats * np.log(rats))))
```

```python
def h(r1, r0, c11, c10, c21, c20, c31, c30):
    totr = r1 + r0
    totc1 = c11 + c10
    totc2 = c21 + c20
    totc3 = c31 + c30
    r1rat = r1 / totr
    r0rat = r0 / totr
    c1rat = totc1 / totr 
    c2rat = totc2 / totr
    c3rat = totc3 / totr
    c11rat = c11 / totc1 if totc1 != 0 else 0
    c10rat = c10 / totc1 if totc1 != 0 else 0
    c21rat = c21 / totc2 if totc2 != 0 else 0
    c20rat = c20 / totc2 if totc2 != 0 else 0
    c31rat = c31 / totc3 if totc3 != 0 else 0
    c30rat = c30 / totc3 if totc3 != 0 else 0
    r1log = log(r1rat) if r1rat != 0 else 0
    r0log = log(r0rat) if r0rat != 0 else 0
    c11log = log(c11rat) if c11rat != 0 else 0
    c10log = log(c10rat) if c10rat != 0 else 0
    c21log = log(c21rat) if c21rat != 0 else 0
    c20log = log(c20rat) if c20rat != 0 else 0
    c31log = log(c31rat) if c31rat != 0 else 0
    c30log = log(c30rat) if c30rat != 0 else 0
   
    return - (r1rat * r1log + r0rat * r0log) + (c1rat*(c11rat*c11log + c10rat*c10log)) + (c2rat*(c21rat*c21log + c20rat*c20log)) + (c3rat*(c31rat*c31log + c30rat*c30log))



```
